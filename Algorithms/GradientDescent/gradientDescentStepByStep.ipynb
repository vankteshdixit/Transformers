```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1, noise=80, random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
plt.show()
```

```python
X
```

```python
y
```

```python
# Apply OLS-> Ordinary least square
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)

LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
plt.show()
```

```python
# Let's apply Gradient Descent assuming slope is constant m = 78.35
# and let's assume the starting value for intercept b=0
```

```python
y_pred = ((78.35 * X) + 0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red', label='OLS')
plt.plot(X,y_pred,color='#00a65a', label='b = 0')
plt.legend()
plt.show()
```

```python
m = 78.35
b = 0

loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope

#ravel-> convert n d array into 1d
```

```python
# Let's take learning rate = 0.1
lr = 0.1

step_size = loss_slope*lr
step_size
```

```python
# Calculating the new intercept
b = b - step_size
b
```

```python
y_pred1 = ((78.35 * X) + b).reshape(4)
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label = 'OLS')
plt.plot(X,y_pred1,color='#00a65a', label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7', label='b = 0')
plt.legend()
```

```python
# Iteration 2
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((78.35 * X) + b).reshape(4)
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label = 'OLS')
plt.plot(X,y_pred2,color='white', label='b = {}'.format(b))
plt.plot(X,y_pred1,color='#00a65a', label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7', label='b = 0')
plt.legend()
```

```python
# Iteration 3
loss_slope = -2 * np.sum(y - m * X.ravel() - b)
loss_slope
step_size = loss_slope * lr
step_size
b = b - step_size
b
y_pred3 = ((78.35 * X) + b).reshape(4)
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red', label='OLS')
plt.plot(X, y_pred3, color='pink', label='b = {}'.format(b))
plt.plot(X, y_pred2, color='white', label='b = {}'.format(b))
plt.plot(X, y_pred1, color='#00a65a', label='b = {}'.format(b))
plt.plot(X, y_pred, color='#A3E4D7', label='b = 0')
plt.legend()
plt.show()
```

```python
# in loop

b = -100
m = 78.35
lr = 0.01

epoches = 100

for i in range(epoches):
     loss_slope = -2 * np.sum(y - m * X.ravel() - b)
     b = b - (lr * loss_slope)

     y_pred = m*X+b
     plt.plot(X,y_pred)

plt.scatter(X,y)
```
