```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
```

```python
df = pd.read_csv('Titanic-Dataset.csv')[['Age','Pclass','SibSp','Parch','Fare','Survived']]
```

```python
# basically feature construction grows with experience
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
X = df.iloc[:,0:4]
y = df.iloc[:,-1]
```

```python
X.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy', cv=20))
```

# Applying Feature Construction

```python
X['Family_size'] = X['SibSp']+X['Parch']+1
```

```python
X.head(1)
```

```python
def myfunc(num):
    if num == 1:
        #alone
        return 0
    elif num > 1 and num <= 4:
        #small family
        return 1
    elif num <= 0:
        #invalid input
        return -1
    else:
        #large family
        return 2
```

```python
myfunc(4)
```

```python
X['Family_type'] = X['Family_size'].apply(myfunc)
```

```python
X.head(1)
```

```python
X.drop(columns=['SibSp','Family_size','Parch'], inplace=True)
```

```python
X.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy', cv=20))
```
